```python
import matplotlib.pyplot as plt
def PieChart(score, labels) :
    fig1 = plt.figure();
    fig1.add_subplot(111)
    plt.pie(score, labels = labels, autopct = '&1.1f%%', startangle = 140)
    plt.axis('equal');
    plt.show()
    return None
```
