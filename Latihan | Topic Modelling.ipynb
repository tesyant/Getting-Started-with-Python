## Word Cloud

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
def wordClouds(teks) :
    wc = WordCloud(background_color = "white") 
    # max_font_size = 40
    wordcloud = wc.generate(teks)
    plt.figure(num = 1, facecolor='w', edgecolor = 'k') 
    # figsize = (4,3), dpi = 600
    # wc.to_file('wordCloud')
    plt.imshow(wordcloud, cmap = plt.cm.jet, interpolation = 'nearest', aspect = 'auto');
    plt.xticks(());
    plt.yticks(())
    plt.show
```

```python
#itcc.wordCloud(' '.join(Doc), plain=True)
wordClouds(text)
```

## Stemming with sastrawi

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stemmer = StemmerFactory().create_stemmer()
```

```python
ktdasar_teks = stemmer.stem(text)
print(ktdasar_teks)
```

```python
wordClouds(ktdasar_teks)
```

## Stopword removal factory

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
```

```python
isiStopWords = factory.get_stop_words()
print(isiStopWords)
```

```python
stopword = factory.create_stop_word_remover()
txtrmvstop = stopword.remove(text)
print(txtrmvstop)
```

```python
wordClouds(txtrmvstop)
```
